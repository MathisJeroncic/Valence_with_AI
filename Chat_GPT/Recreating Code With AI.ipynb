## K-nearest Neighbors Algorithm (k-NN)

In the *first* part of *Valence’s 8th* chapter, the **k-nearest neighbors algorithm** is introduced, a **machine learning algorithm** used for **classification**.

**The original code sample taken from Valence:**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.metrics import accuracy_score
iris = datasets.load_iris()

X = iris.data[:, :2] # only take the first two features
y = iris.target
print("There are",X.shape[0],"data points. Each has",X.shape[1],"features.")

# Create color maps for the three classes:
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

n_neighbors = 15 # define number of neighbors k 
acc = [] # define array to store accuriacies

for weights in ['uniform', 'distance']: # try k-NN for both uniform and distance voting
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    acc.append(accuracy_score(pred,y)) # compare the actual class and the predicted class for all data
    print("There are",accuracy_score(pred,y,normalize=False),"correctly classified samples with",weights,"voting.")

#### The following part is complicated but only concerns the plotting. You don't have to analyze it. ####
    # Plot the results. Assign a color to each point in the coordinate system:
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # range on the x-axis
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # range on the y-axis
    h = .02  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))    
   
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # predict the value of each datapoint in the plot
    
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel("Sepal length")
    plt.ylabel("Sepal width")
    plt.title("3-Class classification (k = %i, weights = '%s')"
              % (n_neighbors, weights))
plt.show()
#### The complicated part about plotting ends here. ####

print("The accuracy when using uniform voting is","{:.2f}".format(acc[0]),".\n")
print("The accuracy when using distance voting is","{:.2f}".format(acc[1]),".\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data[:, :2]  # Use only the first two features (sepal length and sepal width)
y = iris.target

# Define the number of nearest neighbors
n_neighbors = 5

# Create an instance of the KNeighborsClassifier for uniform voting
clf_uniform = KNeighborsClassifier(n_neighbors=n_neighbors, weights='uniform')
clf_uniform.fit(X, y)

# Create an instance of the KNeighborsClassifier for distance voting
clf_distance = KNeighborsClassifier(n_neighbors=n_neighbors, weights='distance')
clf_distance.fit(X, y)

# Create a meshgrid of points to plot the decision boundaries
h = 0.02  # Step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict the class labels for each point on the meshgrid
Z_uniform = clf_uniform.predict(np.c_[xx.ravel(), yy.ravel()])
Z_uniform = Z_uniform.reshape(xx.shape)

Z_distance = clf_distance.predict(np.c_[xx.ravel(), yy.ravel()])
Z_distance = Z_distance.reshape(xx.shape)

# Plot the decision boundaries and the training points
plt.figure(figsize=(10, 6))

# Plot the uniform voting results
plt.subplot(1, 2, 1)
plt.contourf(xx, yy, Z_uniform, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Nearest Neighbors (Uniform Voting)')

# Plot the distance voting results
plt.subplot(1, 2, 2)
plt.contourf(xx, yy, Z_distance, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Nearest Neighbors (Distance Voting)')

plt.tight_layout()
plt.show()
```

## Linear Regression

**Linear regression** is a method in **statistics** used to **train** an algorithm. It’s a part of chapter *8*, part *2.3*.

**The original code sample taken from Valence:**

```python
# Import libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/profit.txt', header=None, names=['Population', 'Profit']) # open data file
print ("Data preview:\n",data.head())
print ("Data statistics:\n",data.describe())

data.plot(kind='scatter', x='Population', y='Profit', figsize=(8,4));   # plot the data points


X = data.iloc[:,0] # the features are contained in all rows, first column
y = data.iloc[:,1] # the values are contained in all rows, second column

# Split the data into traning and test subsets (ratio 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Check the dataset and subsets dimensions: 
print('Dataset size: ', data.shape)
print('X_train size: ', X_train.shape)
print('X_test size: ', X_test.shape)
print('y_train size: ', y_train.shape)
print('y_test size: ', y_test.shape)

# Run the algorithm:
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1,1),y_train) # train
# Get the coefficients: 
theta_0 = lr.intercept_ 
theta_1 = lr.coef_
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1[0]))


y_pred = lr.predict(X_test.values.reshape(-1,1)) # predict
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) # evaluate 


x = np.linspace(data.Population.min(), data.Population.max(), 100)
f = theta_0 + theta_1 * x # this is the linear model for prediction with the parameters theta

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, f, 'r') # plot the prediction line 
ax.scatter(data.Population, data.Profit) # plot the data points
ax.set_xlabel('Population (thousands)')
ax.set_ylabel('Profit (thousands)')
ax.set_title('Predicted Profit vs. Population Size')

prof = lr.predict([[20]]) # add a new town with a population of 20
print("For example, according to our model, a town with a population of 20,000 people would have a weekly profit of","{:.2f}".format(1000*prof[0]),"EUR.")
ax.scatter(20, prof, c='r'); # plot the town and it's profit value on the same figure 
```

## Linear Regression with Multiple Features

In the case of **real-life problems**, the values rarely depend on a **single** feature, as explained in chapter *8*, part *2.4*, so we have asked the chatbots to generate linear regression with **multiple** features.

**The original code sample taken from Valence:**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/prices.txt', header=None, names=['Size', 'Bedrooms', 'Price'])

# Standardize the features in the multiple variable regression:
data_s = (data - data.mean()) / data.std()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data_s['Size'], data_s['Bedrooms'], data_s['Price'], c='c')

X = data_s[['Size', 'Bedrooms']]
y = data_s['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)

lr = LinearRegression()
lr.fit(X_train, y_train)

theta_0 = lr.intercept_
theta_1 = lr.coef_[0]
theta_2 = lr.coef_[1]
print("Theta_0 =", "{:.2f}".format(theta_0))
print("Theta_1 =", "{:.2f}".format(theta_1))
print("Theta_2 =", "{:.2f}".format(theta_2))

y_pred = lr.predict(X_test)
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))

x1 = np.linspace(data_s['Size'].min(), data_s['Size'].max(), 100)
x2 = np.linspace(data_s['Bedrooms'].min(), data_s['Bedrooms'].max(), 100)
x1, x2 = np.meshgrid(x1, x2)
f = theta_0 + theta_1 * x1 + theta_2 * x2

ax.plot_surface(x1, x2, f, alpha=0.5)
ax.set_xlabel('Size (Standardized)')
ax.set_ylabel('Bedrooms (Standardized)')
ax.set_zlabel('Price (Standardized)')
ax.set_title('Predicted Price vs. House Size and Number of Bedrooms')

new_house = data_s.iloc[-1, :]
new_X = new_house[['Size', 'Bedrooms']]
price = lr.predict([new_X])[0]
ax.scatter(new_X['Size'], new_X['Bedrooms'], price, c='r')

plt.show()
```

4. Can you visualise it in 3D?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Read the data from the txt file
data_url = "https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/prices.txt"
data = pd.read_csv(data_url, header=None, names=['Size', 'Bedrooms', 'Price'])

# Extract the features and target variables
X = data[['Size', 'Bedrooms']]
y = data['Price']

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Generate predictions for all combinations of Size and Bedrooms
x1_range = np.linspace(min(X['Size']), max(X['Size']), 100)
x2_range = np.linspace(min(X['Bedrooms']), max(X['Bedrooms']), 100)
X_pred = np.array(np.meshgrid(x1_range, x2_range)).T.reshape(-1, 2)
y_pred = model.predict(X_pred)

# Visualize the data in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X['Size'], X['Bedrooms'], y, c='blue', label='Actual Price')
ax.scatter(X_pred[:, 0], X_pred[:, 1], y_pred, c='red', label='Predicted Price')
ax.set_xlabel('Size (sqft)')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('Price (EUR)')
ax.set_title('House Price Prediction')
ax.legend()
plt.show()
```

**Final result:**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Read the data from the txt file
data_url = "https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/prices.txt"
data = pd.read_csv(data_url, header=None, names=['Size', 'Bedrooms', 'Price'])

# Extract the features and target variables
X = data[['Size', 'Bedrooms']]
y = data['Price']

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a linear regression model
model = LinearRegression()

# Train the model
model.fit(X_scaled, y)

# Generate predictions for all combinations of Size and Bedrooms
x1_range = np.linspace(min(X_scaled[:, 0]), max(X_scaled[:, 0]), 100)
x2_range = np.linspace(min(X_scaled[:, 1]), max(X_scaled[:, 1]), 100)
X_pred_scaled = np.array(np.meshgrid(x1_range, x2_range)).T.reshape(-1, 2)
X_pred = scaler.inverse_transform(X_pred_scaled)
y_pred = model.predict(X_pred_scaled)

# Visualize the data in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X['Size'], X['Bedrooms'], y, c='blue', label='Actual Price')
ax.scatter(X_pred[:, 0], X_pred[:, 1], y_pred, c='red', alpha=0.02, label='Predicted Price')  # Setting alpha for transparency
ax.set_xlabel('Size (sqft)')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('Price (EUR)')
ax.set_title('House Price Prediction')
ax.legend()
plt.show()
```
