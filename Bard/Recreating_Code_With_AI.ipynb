## K-nearest Neighbors Algorithm (k-NN)

In the *first* part of *Valence’s 8th* chapter, the **k-nearest neighbors algorithm** is introduced, a **machine learning algorithm** used for **classification**.

**The original code sample taken from Valence:**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.metrics import accuracy_score
iris = datasets.load_iris()

X = iris.data[:, :2] # only take the first two features
y = iris.target
print("There are",X.shape[0],"data points. Each has",X.shape[1],"features.")

# Create color maps for the three classes:
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

n_neighbors = 15 # define number of neighbors k 
acc = [] # define array to store accuriacies

for weights in ['uniform', 'distance']: # try k-NN for both uniform and distance voting
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights) # create a Neighbours Classifier
    clf.fit(X, y) # fit the data
    pred = clf.predict(X) # predict values for all data
    acc.append(accuracy_score(pred,y)) # compare the actual class and the predicted class for all data
    print("There are",accuracy_score(pred,y,normalize=False),"correctly classified samples with",weights,"voting.")

#### The following part is complicated but only concerns the plotting. You don't have to analyze it. ####
    # Plot the results. Assign a color to each point in the coordinate system:
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1 # range on the x-axis
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1 # range on the y-axis
    h = .02  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))    
   
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # predict the value of each datapoint in the plot
    
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xlabel("Sepal length")
    plt.ylabel("Sepal width")
    plt.title("3-Class classification (k = %i, weights = '%s')"
              % (n_neighbors, weights))
plt.show()
#### The complicated part about plotting ends here. ####

print("The accuracy when using uniform voting is","{:.2f}".format(acc[0]),".\n")
print("The accuracy when using distance voting is","{:.2f}".format(acc[1]),".\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the Iris dataset
data = np.loadtxt("iris.csv", delimiter=",")

# Split the data into features and labels
features = data[:, :2]
labels = data[:, -1]

# Create color maps for the three classes of flowers
colors = ["red", "green", "blue"]

# Create a function to calculate the distance between two points
def distance(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Create a function to classify a point using the k-nearest neighbors algorithm
def classify(point, k, distance_metric="euclidean"):
    # Find the k nearest neighbors of the point
    neighbors = np.argsort([distance(point, p) for p in features])[:k]

    # Get the labels of the k nearest neighbors
    neighbor_labels = labels[neighbors]

    # If using uniform voting, simply count the number of votes for each class
    if distance_metric == "uniform":
        votes = np.bincount(neighbor_labels)

    # If using distance voting, calculate the weighted votes for each class
    else:
        votes = np.zeros(len(np.unique(labels)))
        for label in neighbor_labels:
            votes[label] += 1 / distance(point, features[neighbors[label]])

    # Return the class with the most votes
    return np.argmax(votes)

# Classify the points using the k-nearest neighbors algorithm
predictions = []
for point in features:
    predictions.append(classify(point, 5))

# Plot the results
plt.scatter(features[:, 0], features[:, 1], c=colors[labels], alpha=0.5)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.title("Iris dataset classified using k-nearest neighbors")
plt.show()
```

## Linear Regression

**Linear regression** is a method in **statistics** used to **train** an algorithm. It’s a part of chapter *8*, part *2.3*.

**The original code sample taken from Valence:**

```python
# Import libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('https://raw.githubusercontent.com/VALENCEML/eBOOK/main/EN/08/profit.txt', header=None, names=['Population', 'Profit']) # open data file
print ("Data preview:\n",data.head())
print ("Data statistics:\n",data.describe())

data.plot(kind='scatter', x='Population', y='Profit', figsize=(8,4));   # plot the data points


X = data.iloc[:,0] # the features are contained in all rows, first column
y = data.iloc[:,1] # the values are contained in all rows, second column

# Split the data into traning and test subsets (ratio 9:1): 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/10, random_state=0) 
# Check the dataset and subsets dimensions: 
print('Dataset size: ', data.shape)
print('X_train size: ', X_train.shape)
print('X_test size: ', X_test.shape)
print('y_train size: ', y_train.shape)
print('y_test size: ', y_test.shape)

# Run the algorithm:
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1,1),y_train) # train
# Get the coefficients: 
theta_0 = lr.intercept_ 
theta_1 = lr.coef_
print("Theta_0=","{:.2f}".format(theta_0))
print("Theta_1=","{:.2f}".format(theta_1[0]))


y_pred = lr.predict(X_test.values.reshape(-1,1)) # predict
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred)) # evaluate 


x = np.linspace(data.Population.min(), data.Population.max(), 100)
f = theta_0 + theta_1 * x # this is the linear model for prediction with the parameters theta

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(x, f, 'r') # plot the prediction line 
ax.scatter(data.Population, data.Profit) # plot the data points
ax.set_xlabel('Population (thousands)')
ax.set_ylabel('Profit (thousands)')
ax.set_title('Predicted Profit vs. Population Size')

prof = lr.predict([[20]]) # add a new town with a population of 20
print("For example, according to our model, a town with a population of 20,000 people would have a weekly profit of","{:.2f}".format(1000*prof[0]),"EUR.")
ax.scatter(20, prof, c='r'); # plot the town and it's profit value on the same figure 
```

## DBSCAN

In some cases KMeans does not perform very well, so we can use DBSCAN (aka. Density-based spatial clustering of applications with noise) algorithm instead, which is also a part of *9th* in chapter.

**The original code sample taken from Valence:**

```python
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

from sklearn.cluster import DBSCAN 

n_samples = 1500 # number of data points
random_state = 100 # seed number for data points generation 
# Generate concentrical circles:
noisy_circles = datasets.make_circles(n_samples=n_samples,factor=.5,noise=.05)
# Generate crescents:
noisy_moons = datasets.make_moons(n_samples=n_samples,noise=.05) 

# Plot the generated datasets:
datsets2 = {'circles': noisy_circles,'crescents': noisy_moons}
plot_nr = 1
plt.figure(figsize=(20, 5))

plt.figure(figsize=(30, 10)) # generate an empty figure
plot_nr = 1 # start a counter for the number of plots

for key in datsets2: # for each dataset:
    X, y = datsets2[key] # retrieve data
    scaler = StandardScaler() # standardize data
    dbscan = DBSCAN(eps=0.3,min_samples=10) # initialize algorithm
    y_pred = dbscan.fit_predict(scaler.fit_transform(X)) # run algorithm
    # plot:
    plt.subplot(1, len(datsets2), plot_nr)
    plt.scatter(X[:, 0], X[:, 1], c=y_pred) 
    plt.xlabel('Feature 1 - X1 coordinate')
    plt.ylabel('Feature 2 - X2 coordinate')
    plot_nr += 1
```

## Implementing neuron training

To be able to train our neuron we will implement a function `backprop()` to calculate the gradient of the loss function with respect to the two weights and the bias.
We will also implement an `update()` function to change the neuron parameters according to the calculated gradient and the learning rate.
Finally, we'll implement a function `fit()` that we can call to train the neuron.

**The original code sample taken from Valence:**

```python
class Neuron:
    def __init__(self, weights=None, bias=None, n_inputs=None):
        if weights is not None:
            self.weights = weights
            self.bias = bias
        else:
            self.weights = np.random.normal(size=n_inputs)
            self.bias = np.random.normal()

    def __call__(self, x):
        a = np.sum(self.weights * x) + self.bias
        y = 1 if a > 0 else 0
        return y

    def backprop(self, loss, x):
        # loss = y - y_pred = y - (sum(w * x) + b)
        self.grad_weights = - loss * x
        self.grad_bias = - loss

    def update(self):
        self.weights = self.weights - self.grad_weights * self.learn_rate
        self.bias = self.bias - self.grad_bias * self.learn_rate

    def fit(self, xs, ys, epochs=20, learn_rate=0.1, verbose=True):
        self.learn_rate = learn_rate
        for epoch in range(epochs):
            if verbose:
                print(f"{epoch + 1}/{epochs} loss = ", end="")
            epoch_loss = 0
            for x, y in zip(xs, ys):
                y_pred = self(x)
                loss = y - y_pred
                self.backprop(loss, x)
                self.update()
                epoch_loss += loss
            if verbose:
                print(epoch_loss / x.shape[0])

np.random.seed(42)
neuron = Neuron(n_inputs=2)
print(neuron.weights, neuron.bias)
for x in xs:
    print(x, "->", neuron(x))

ys = np.array([0, 0, 0, 1])  # AND

neuron.fit(xs, ys)

print(neuron.weights, neuron.bias)
for x in xs:
    print(x, "->", neuron(x))
```

```python
class Neuron:

    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def call(self, inputs):
        """
        Calculates the output of the neuron given a set of inputs.

        Args:
            inputs: A list of inputs.

        Returns:
            The output of the neuron.
        """
        output = 0
        for i in range(len(inputs)):
            output += inputs[i] * self.weights[i]
        output += self.bias
        return output

    def backprop(self, inputs, output, target):
        """
        Calculates the gradient of the loss function with respect to the two weights and the bias.

        Args:
            inputs: A list of inputs.
            output: The output of the neuron.
            target: The target value.

        Returns:
            A tuple containing the gradients of the loss function with respect to the two weights and the bias.
        """
        gradients = []
        for i in range(len(inputs)):
            gradients.append(output * inputs[i])
        gradients.append(output)
        return gradients

    def update(self, gradients, learning_rate):
        """
        Updates the neuron parameters according to the calculated gradient and the learning rate.

        Args:
            gradients: A tuple containing the gradients of the loss function with respect to the two weights and the bias.
            learning_rate: The learning rate.
        """
        for i in range(len(self.weights)):
            self.weights[i] -= gradients[i] * learning_rate
        self.bias -= gradients[-1] * learning_rate

    def fit(self, inputs, targets, epochs, learning_rate):
        """
        Trains the neuron using the given inputs and targets.

        Args:
            inputs: A list of inputs.
            targets: A list of target values.
            epochs: The number of epochs to train for.
            learning_rate: The learning rate.
        """
        for epoch in range(epochs):
            for i in range(len(inputs)):
                output = self.call(inputs[i])
                gradients = self.backprop(inputs[i], output, targets[i])
                self.update(gradients, learning_rate)

```

```python
import numpy as np

class Neuron:

    def __init__(self, weights, bias):
        self.weights = weights
        self.bias = bias

    def call(self, inputs):
        return np.dot(inputs, self.weights) + self.bias

    def backprop(self, loss, inputs):
        gradients = np.dot(inputs.T, loss)
        return gradients

    def update(self, gradients, learning_rate):
        self.weights -= learning_rate * gradients
        self.bias -= learning_rate * gradients[0]

    def fit(self, inputs, targets, learning_rate, epochs):
        for epoch in range(epochs):
            loss = 0
            for i in range(len(inputs)):
                output = self.call(inputs[i])
                loss += (targets[i] - output) ** 2
                gradients = self.backprop(loss, inputs[i])
                self.update(gradients, learning_rate)
            print(f"Error at epoch {epoch}: {loss}")

weights = np.array([1, 2, 3])
bias = 4

neuron = Neuron(weights, bias)

inputs = np.array([1, 2, 3])
output = neuron.call(inputs)
print(output)

```

## Regression with a Shallow Neural Network

Let's make a neural network to predict the output of a sine function for input data that is not part of the training set.
To work with neural networks we will use the `scikit-learn` package.

**The original code sample taken from Valence:**

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import neural_network
xs = np.linspace(0, 2 * np.pi, 10)
np.random.seed(42)
ys = np.sin(xs) + np.random.normal(size=xs.size) * 0.2
x_axis = np.linspace(-.1, 2 * np.pi + .1, 100)
y_sin = np.sin(x_axis)

xs = np.expand_dims(xs, 1)
x_axis = np.expand_dims(x_axis, 1)

##show the resulting training set and the sine function
plt.figure()
plt.scatter(xs, ys)
plt.plot(x_axis, y_sin, ':', lw=4, alpha=0.2)
plt.grid()

##initialize the neural network
reg = neural_network.MLPRegressor(
    hidden_layer_sizes=(5),
    activation="tanh",
    solver="adam",
    alpha=0.0001,
    batch_size="auto",
    learning_rate_init=0.01,
    max_iter=50,
    tol=1e-4,
    early_stopping=False,
    random_state=42,
    verbose=True,    
    )
##train the network
reg.fit(xs, ys)

##predict values for the learned function
y_pred_under = reg.predict(x_axis)
##show result
plt.figure()
plt.scatter(xs, ys, c='k')
plt.plot(x_axis, y_sin, ':', lw=4, alpha=0.2)
plt.plot(x_axis, y_pred_under)
plt.grid()
plt.tight_layout()
```
